```python
numbers = [10, 20, 30, 40, 50]

n = int(input("Enter the number to search: "))

if n in numbers:
    position = numbers.index(n)
    print(f"Number {n} is present at position {position}")
else:
    print(f"Number {n} is not present in the list")
```

```python
numbers = [10, -5, 20, -15, 0, 25, -30]

positive = []
negative = []

for num in numbers:
    if num > 0:
        positive.append(num)
    elif num < 0:
        negative.append(num)

print("Original list:", numbers)
print("Positive numbers:", positive)
print("Negative numbers:", negative)
```
