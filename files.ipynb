```python


filename = "student.txt"


with open(filename, "w") as file:
    sid = input("Enter Student ID: ")
    name = input("Enter Name: ")
    roll = input("Enter Roll No: ")
    branch = input("Enter Branch: ")
    marks = input("Enter Marks: ")
    address = input("Enter Address: ")

    file.write(f"ID: {sid}\n")
    file.write(f"Name: {name}\n")
    file.write(f"Roll No: {roll}\n")
    file.write(f"Branch: {branch}\n")
    file.write(f"Marks: {marks}\n")
    file.write(f"Address: {address}\n")

print("\nStudent data written to file successfully.\n")

print("Reading data from file:\n")
with open(filename, "r") as file:
    print(file.read())
```

```python
import pickle

filename = "student.dat"


with open(filename, "wb") as file:
    student = {}
    student["stu_id"] = input("Enter Student ID: ")
    student["name"] = input("Enter Name: ")
    student["year"] = input("Enter Year: ")
    student["branch"] = input("Enter Branch: ")
    student["marks"] = float(input("Enter Marks: "))

    pickle.dump(student, file)

print("\nStudent record written to binary file.\n")


print("Reading data from binary file:\n")
with open(filename, "rb") as file:
    data = pickle.load(file)
    for key, value in data.items():
        print(f"{key} : {value}")
```
