```python
courses = {"CSE", "ECE", "EEE", "MECH", "CIVIL"}

print("Courses offered:", courses)
print("Number of courses:", len(courses))

branch = input("Enter branch to check: ")

if branch in courses:
    print(branch, "is available")
else:
    print(branch, "is not available")
```

```python
odd = {1, 3, 5, 7, 9}
even = {2, 4, 6, 8, 10}

merged = odd.union(even)

print("Odd numbers:", odd)
print("Even numbers:", even)
print("Merged set:", merged)
```
