```python
class Student:
    def __init__(self, sid, name, branch, marks):
        self.sid = sid
        self.name = name
        self.branch = branch
        self.marks = marks

    def percentage(self):
        return sum(self.marks) / len(self.marks)

    def grade(self):
        p = self.percentage()
        if p >= 75:
            return "A"
        elif p >= 60:
            return "B"
        elif p >= 50:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("ID:", self.sid)
        print("Name:", self.name)
        print("Branch:", self.branch)
        print("Percentage:", self.percentage())
        print("Grade:", self.grade())
        print("--------------------")



s1 = Student(101, "Sumanth", "CSE", [80, 75, 70, 85, 90])
s2 = Student(102, "Ravi", "ECE", [65, 60, 58, 62, 68])


s1.display()
s2.display()
```

```python
class RGUKT:
    def __init__(self, location, strength, courses):
        self.location = location
        self.strength = strength
        self.courses = courses

    def display(self):
        print("Location:", self.location)
        print("Student Strength:", self.strength)
        print("Courses Offered:", self.courses)
        print("--------------------")



rkv = RGUKT("RK Valley", 12000, ["PUC", "B.Tech"])
ongole = RGUKT("Ongole", 8000, ["PUC", "B.Tech"])


rkv.display()
ongole.display()
```
