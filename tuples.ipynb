```python
n = int(input("Enter number of students: "))
emails = []

for i in range(n):
    mail = input(f"Enter email id {i+1}: ")
    emails.append(mail)

email_tuple = tuple(emails)

usernames = tuple(mail.split("@")[0] for mail in email_tuple)
domains = tuple(mail.split("@")[1] for mail in email_tuple)

print("\nEmail IDs:", email_tuple)
print("Usernames:", usernames)
print("Domains:", domains)
```

```python
n = int(input("Enter number of students: "))
names = []

for i in range(n):
    names.append(input("Enter student name: "))

students = tuple(names)

search = input("Enter name to search: ")

if search in students:
    print(search, "is present in the tuple")
else:
    print(search, "is not present in the tuple")
```
